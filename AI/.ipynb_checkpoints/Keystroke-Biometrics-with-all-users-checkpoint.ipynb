```python
import os
import pandas as pd
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.neural_network import MLPClassifier
from sklearn import svm
from sklearn.metrics import f1_score
from sklearn.svm import SVR
from keras.models import Sequential
from keras.layers import Dense
import seaborn as sns
from sklearn.metrics import accuracy_score
```

Random Forest Classifier Model Training and Attack

```python
def rfc_model(X_train,X_test,y_train,y_test,X1,y2):
    rfc = RandomForestClassifier(n_estimators=1000)
    rfc.fit(X_train, y_train)
    pred_rfc = rfc.predict(X_test)
    print("\n RFC:")
    print(classification_report(y_test, pred_rfc))
    print(confusion_matrix(y_test, pred_rfc))
    print("\n RFC Attack:")
    pred_rfc2 = rfc.predict(X1)
    print(classification_report(y2, pred_rfc2))
    print(confusion_matrix(y2, pred_rfc2))
```

SVM Model Training and Attack

```python
def svm_model(X_train,X_test,y_train,y_test,X1,y2):
    clf = svm.SVC()
    clf.fit(X_train, y_train)
    pred_clf = clf.predict(X_test)
    print("\n SVM:")
    print(classification_report(y_test, pred_clf))
    print(confusion_matrix(y_test, pred_clf))
    print("\n SVM Attack:")
    pred_clf2 = clf.predict(X1)
    print(classification_report(y2, pred_clf2))
    print(confusion_matrix(y2, pred_clf2))
```

Neural Net Model Training and Attack

```python
def mlpc_model(X_train,X_test,y_train,y_test,X1,y2):
    mlpc=MLPClassifier(hidden_layer_sizes=(20,20,20,20),max_iter=2000)
    mlpc.fit(X_train, y_train)
    pred_mlpc = mlpc.predict(X_test)
    print("\n mlpc:")
    print(classification_report(y_test, pred_mlpc))
    print(confusion_matrix(y_test, pred_mlpc))
    print("\n MLPC Attack:")
    pred_mlpc2 = mlpc.predict(X1)
    print(classification_report(y2, pred_mlpc2))
    print(confusion_matrix(y2, pred_mlpc2))
```

Taking CSV data for Training, Testing and Attacking and feeding it into the Model Methods

```python
def data_split(train_test_file,attack_file):
    train_csv = train_test_file[0]
    test_csv = train_test_file[1]
    
    train_set = pd.read_csv("train_test_attack_v2/TrainTest/"+str(train_csv))
    test_set = pd.read_csv("train_test_attack_v2/TrainTest/"+str(test_csv))
    attack_set = pd.read_csv("train_test_attack_v2/Attack/"+str(attack_file))
    
    X_train = train_set.drop('uid', axis=1)
    X_test = test_set.drop('uid', axis=1)
    y_train = train_set['uid']
    y_test = test_set['uid']
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    X1 = attack_set.drop('uid', axis=1)
    y2 = attack_set['uid']
    X1 = sc.transform(X1)
    
    rfc_model(X_train,X_test,y_train,y_test,X1,y2)
    svm_model(X_train,X_test,y_train,y_test,X1,y2)
    mlpc_model(X_train,X_test,y_train,y_test,X1,y2)
```

Reading the Data From the Directories

```python
files = os.listdir("train_test_attack_v2/TrainTest")
tuples = []
for num in range(1,31):
    sub_a = "a"+str(num)+"-"
    sub_b = "b"+str(num)+"-"
    file_a = ""
    file_b = ""
    for file in files:
        if sub_a in file:
            file_a = file
        if sub_b in file:
            file_b = file
    new_tuple = file_a,file_b
    tuples.append(new_tuple)

attack_files = os.listdir("train_test_attack_v2/Attack")
sorted_attack_files = []
for num in range(1,31):
    sub_a = "Attack"+str(num)+"with"
    file_a = ""
    for file in attack_files:
        if sub_a in file:
            sorted_attack_files.append(file)
```

Running the Model and Displaying the Results

```python
classification_models = []

for num in range(0,30):
    print(str(tuples[num])+"\n"+str(sorted_attack_files[num]))
    data_split(tuples[num], sorted_attack_files[num])
    
```
