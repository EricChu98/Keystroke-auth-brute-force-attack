```python
import os
import pandas as pd
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.neural_network import MLPClassifier
from sklearn import svm


def data_ag(folder, file_name):
    stringer = folder + "/"+file_name
    with open(stringer) as my_file:
        testsite_array = my_file.readlines()
    my_file.close()

    for line in testsite_array:
        if line == '\n':
            testsite_array.remove(line)

    remove_elem1 = testsite_array[0]
    remove_elem2 = testsite_array[len(testsite_array)-1]
    new_testsite_array = []
    new_testsite_array.append(testsite_array[1])
    new_testsite_array.append(testsite_array[2])
    new_testsite_array.append(testsite_array[3])
    new_testsite_array.append(testsite_array[4])
    
    new_array = []
    final_array = []
    for line in new_testsite_array:
        new_array.append(line.split(','))
    new_count = len(new_array[0])
    index = []
    for num in range(0,new_count):
        index.append(int(folder))
    final_array.append(index)
    for array in new_array:
        array[len(array)-1] = array[len(array)-1].strip('\n')
        desired_array = [int(numeric_string) for numeric_string in array]
        final_array.append(desired_array)
    final_array[0].append('xxx')
    final_array[1].append('xxx')
    final_array[2].append('xxx')
    final_array[3].append('xxx')
    final_array[4].append('xxx')
 #   print(str(len(final_array[0]))+ "\t"+str(len(final_array[1]))+"\t"+str(len(final_array[2]))+"\t"+str(len(final_array[3]))+"\t"+str(len(final_array[4])))
    df = pd.DataFrame({'uid': final_array[0], 'press_down_id': final_array[1], 'press_down_time': final_array[2]})
    df2 = pd.DataFrame({'release_id': final_array[3], 'key_release_time': final_array[4]})
    result = pd.concat([df,df2], axis=1)
    return result
```

```python

dir_list = os.listdir()
for directory in dir_list:
    if "0" not in str(directory):
        dir_list.remove(directory)
dir_list.remove(dir_list[len(dir_list) - 1])

new_dir_dict = {}

for directory in dir_list:
    sub_dir = os.listdir(str(directory))
    for file in sub_dir:
        if ".txt" not in str(file):
            sub_dir.remove(file)
    new_dir_dict[str(directory)] = sub_dir

keys = new_dir_dict.keys()
new_keys = []
for key in keys:
    new_keys.append(key)

uid = []
press_down_id = []
press_down_time = []
release_id = []
key_release_time = []
newDF = pd.DataFrame()
for key in new_keys:
    for file_name in new_dir_dict[key]:
        newDF = newDF.append(data_ag(key,file_name))
```

```python
newDF.info()
newDF = newDF.fillna(0)
newDF.to_csv('data.csv', index=False)

```

```python
X = newDF.drop('uid', axis=1)
y = newDF['uid']
```

```python
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 42)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
pred_rfc = rfc.predict(X_test)
```

```python
print(classification_report(y_test, pred_rfc))
print(confusion_matrix(y_test, pred_rfc))
```
